```python
import cv2

from ultralytics import YOLO
from collections import defaultdict
import numpy as np
```

```python
# Load the YOLO11 model
model = YOLO("yolo11n.pt")
```

```python
# Open the video file
video_path = "default_dataset.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
results = model.track(source=video_path,show=True,tracker="bytetrack.yaml",classes=[41])
```

```python
with open("results.txt", "w") as outfile:
    frame_idx = 1  # l'index de frame doit dbuter  1
    for result in results:
        if hasattr(result, "boxes") and result.boxes is not None:
            for box in result.boxes:
                # Rcuprer les coordonnes sous le format [xmin, ymin, xmax, ymax]
                xyxy = box.xyxy.cpu().numpy()[0]
                xmin, ymin, xmax, ymax = map(int, xyxy)
                width = xmax - xmin
                height = ymax - ymin

                # Gestion de l'identifiant du suivi en vrifiant qu'il n'est pas None
                if hasattr(box, "id") and box.id is not None:
                    tracker_id = int(box.id)
                elif hasattr(box, "tracker_id") and box.tracker_id is not None:
                    tracker_id = int(box.tracker_id)
                else:
                    tracker_id = 0  # Valeur par dfaut si aucun identifiant n'est disponible

                outfile.write(f"{frame_idx} {tracker_id} {xmin} {ymin} {width} {height}\n")
        frame_idx += 1

print("Le fichier results.txt a t cr avec succs.")
```
